```python
from matplotlib.patches import Circle, Rectangle, Arc
# 1 foot = 10 Aribitrary units

def draw_half_court(ax=None, color='black', lw=2, outer_lines=False):
    # If an axes object isn't provided to plot onto, just get current one
    if ax is None:
        ax = plt.gca()

    # Create the various parts of an NBA basketball court

    # Create the basketball hoop
    # Diameter of a hoop is 18" so it has a radius of 9", which is a value
    # 7.5 in our coordinate system
    hoop = Circle((0, 0), radius=7.5, linewidth=lw, color=color, fill=False)

    # Create backboard
    backboard = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color=color)

    # The paint
    # Create the outer box 0f the paint, width=16ft, height=19ft
    outer_box = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color=color,
                          fill=False)
    # Create the inner box of the paint, widt=12ft, height=19ft
    inner_box = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color=color,
                          fill=False)

    # Create free throw top arc
    top_free_throw = Arc((0, 142.5), 120, 120, theta1=0, theta2=180,
                         linewidth=lw, color=color, fill=False)
    # Create free throw bottom arc
    bottom_free_throw = Arc((0, 142.5), 120, 120, theta1=180, theta2=0,
                            linewidth=lw, color=color, linestyle='dashed')
    # Restricted Zone, it is an arc with 4ft radius from center of the hoop
    restricted = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw,
                     color=color)

    # Three point line
    # Create the side 3pt lines, they are 14ft long before they begin to arc
    corner_three_a = Rectangle((-220, -47.5), 0, 140, linewidth=lw,
                               color=color)
    corner_three_b = Rectangle((220, -47.5), 0, 140, linewidth=lw, color=color)
    # 3pt arc - center of arc will be the hoop, arc is 23'9" away from hoop
    # I just played around with the theta values until they lined up with the 
    # threes
    three_arc = Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=lw,
                    color=color)

    # Center Court
    center_outer_arc = Arc((0, 422.5), 120, 120, theta1=180, theta2=0,
                           linewidth=lw, color=color)
    center_inner_arc = Arc((0, 422.5), 40, 40, theta1=180, theta2=0,
                           linewidth=lw, color=color)

    # List of the court elements to be plotted onto the axes
    court_elements = [hoop, backboard, outer_box, inner_box, top_free_throw,
                      bottom_free_throw, restricted, corner_three_a,
                      corner_three_b, three_arc, center_outer_arc,
                      center_inner_arc]

    if outer_lines:
        # Draw the half court line, baseline and side out bound lines
        outer_lines = Rectangle((-250, -47.5), 500, 470, linewidth=lw,
                                color=color, fill=False)
        court_elements.append(outer_lines)

    # Add the court elements onto the axes
    for element in court_elements:
        ax.add_patch(element)

    return ax

def draw_full_court(ax=None, color='black', lw=2, outer_lines=False):
    # If an axes object isn't provided to plot onto, just get current one
    if ax is None:
        ax = plt.gca()

    # Create the various parts of an NBA basketball court

    # Create the basketball hoop
    # Diameter of a hoop is 18" so it has a radius of 9", which is a value
    # 7.5 in our coordinate system
    hoop = Circle((0, 0), radius=7.5, linewidth=lw, color=color, fill=False)
    hoop2 = Circle((0, 845), radius=7.5, linewidth=lw, color=color, fill=False)


    # Create backboard
    backboard = Rectangle((-30, -7.5), 60, -1, linewidth=lw, color=color)
    backboard2 = Rectangle((-30, 852), 60, -1, linewidth=lw, color=color)


    # The paint
    # Create the outer box 0f the paint, width=16ft, height=19ft
    outer_box = Rectangle((-80, -47.5), 160, 190, linewidth=lw, color=color, fill=False)
    outer_box2 = Rectangle((-80, 892.5), 160, -190, linewidth=lw, color=color, fill=False)

    # Create the inner box of the paint, widt=12ft, height=19ft
    inner_box = Rectangle((-60, -47.5), 120, 190, linewidth=lw, color=color, fill=False)
    inner_box2 = Rectangle((-60, 892.5), 120, -190, linewidth=lw, color=color, fill=False)


    # Create free throw top arc
    top_free_throw = Arc((0, 142.5), 120, 120, theta1=0, theta2=180, linewidth=lw, color=color, fill=False)
    top_free_throw2 = Arc((0, 702.5), 120, 120, theta1=180, theta2=360, linewidth=lw, color=color, fill=False)

    # Create free throw bottom arc
    bottom_free_throw = Arc((0, 142.5), 120, 120, theta1=180, theta2=0,linewidth=lw, color=color, linestyle='dashed')
    bottom_free_throw2 = Arc((0, 702.5), 120, 120, theta1=360, theta2=180,linewidth=lw, color=color, linestyle='dashed')

    # Restricted Zone, it is an arc with 4ft radius from center of the hoop
    restricted = Arc((0, 0), 80, 80, theta1=0, theta2=180, linewidth=lw, color=color)
    restricted2 = Arc((0, 852), 80, 80, theta1=180, theta2=360, linewidth=lw, color=color)


    # Three point line
    # Create the side 3pt lines, they are 14ft long before they begin to arc
    corner_three_a = Rectangle((-220, -47.5), 0, 140, linewidth=lw, color=color)
    corner_three_a2 = Rectangle((-220, 892.5), 0, -140, linewidth=lw, color=color)

    corner_three_b = Rectangle((220, -47.5), 0, 140, linewidth=lw, color=color)
    corner_three_b2 = Rectangle((220, 892.5), 0, -140, linewidth=lw, color=color)

    # 3pt arc - center of arc will be the hoop, arc is 23'9" away from hoop
    # I just played around with the theta values until they lined up with the 
    # threes
    three_arc = Arc((0, 0), 475, 475, theta1=22, theta2=158, linewidth=lw,color=color)
    three_arc2 = Arc((0, 845), 475, 475, theta1=202, theta2=338, linewidth=lw,color=color)


    
    # Center Court
    center_outer_arc = Circle((0, 422.5), radius=60, linewidth=lw, color=color, fill=False)
    
    center_inner_arc = Circle((0, 422.5), radius=20, linewidth=lw, color=color, fill=False)

    # List of the court elements to be plotted onto the axes
    court_elements = [hoop, hoop2, backboard, backboard2, outer_box, outer_box2, 
                      inner_box, inner_box2, top_free_throw,top_free_throw2, 
                      bottom_free_throw, bottom_free_throw2, restricted, restricted2, 
                      corner_three_a,corner_three_a2, corner_three_b, corner_three_b2, 
                      three_arc, three_arc2, center_outer_arc, center_inner_arc]

    if outer_lines:
        # Draw the half court line, baseline and side out bound lines
        outer_lines = Rectangle((-250, -47.5), 500, 940, linewidth=lw, color=color, fill=False)
        center_court_line = Rectangle((-250, 422.5), 500, 0, linewidth=lw, color=color, fill=False)

        court_elements.append(outer_lines)
        court_elements.append(center_court_line)


    # Add the court elements onto the axes
    for element in court_elements:
        ax.add_patch(element)

    return ax
```

```python
plt.figure(figsize=(10,19))
draw_full_court(outer_lines=True)
plt.xlim(-251,251)
plt.ylim(-100,1100)
plt.show()
```

```python
plt.figure(figsize=(14,19))
plt.scatter(df.LOC_X, df.LOC_Y)
draw_full_court(outer_lines=True)
# Descending values along the axis from left to right
plt.xlim(300,-300)
plt.ylim(-50,900)
plt.show()
```

```python
plt.figure(figsize=(14,19))
df = df[df.SHOT_MADE_FLAG == 1]

plt.scatter(df.LOC_X, df.LOC_Y)
draw_full_court()
# Adjust plot limits to just fit in half court
plt.xlim(-270,270)
# Descending values along th y axis from bottom to top
# in order to place the hoop by the top of plot
plt.ylim(900, -47.5)
# get rid of axis tick labels
plt.tick_params(labelbottom=False, labelleft=False)
plt.show()
```

```python
# create our jointplot
cmap=plt.cm.YlOrRd_r 

joint_shot_chart = sns.jointplot(df.LOC_X, df.LOC_Y, stat_func=None,
                                 kind='kde', space=0, alpha=0.5, color = cmap(0.4), cmap = cmap, n_levels=100)

joint_shot_chart.fig.set_size_inches(14,19)

# A joint plot has 3 Axes, the first one called ax_joint 
# is the one we want to draw our court onto and adjust some other settings
ax = joint_shot_chart.ax_joint
draw_full_court(ax)

# Adjust the axis limits and orientation of the plot in order
# to plot half court, with the hoop by the top of the plot
ax.set_xlim(-270,270)
ax.set_ylim(900, -47.5)

# Get rid of axis labels and tick marks
ax.set_xlabel('')
ax.set_ylabel('')
ax.tick_params(labelbottom='off', labelleft='off')

# Add a title
ax.set_title('Stephen Curry FGA \n2015-16 Reg. Season', 
             y=1.2, fontsize=18)

# Add Data Scource and Author
ax.text(-250,445,'Data Source: stats.nba.com',
        fontsize=12)

plt.show()
```
